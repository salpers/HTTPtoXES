```python
from event_loop.preprocessing.event_attributes import extract_applicant_id, extract_activity_id, extract_mail_id

%autoreload 2

df_train_pp["applicant_id"] = df_train_pp.apply(lambda x: extract_applicant_id(x["pgsql.query"], x["pgsql.query_0"], x["pgsql.target"], x["file_data"], x["selective_file_data"], x["origin_file_data"]), axis = 1)


df_train_pp["activity_id"] = df_train_pp.apply(lambda x: extract_activity_id(x["pgsql.query"], x["pgsql.query_0"], x["pgsql.target"], x["file_data"], x["selective_file_data"], x["origin_file_data"]), axis = 1)

df_train_pp["mail_id"] = df_train_pp.apply(lambda x: extract_mail_id(x["pgsql.query"], x["pgsql.query_0"], x["pgsql.target"], x["file_data"], x["selective_file_data"], x["origin_file_data"]), axis = 1)
```

```python
df_train_pp.groupby(["BusinessActivity"]).agg({"applicant_id": "unique", "activity_id": "unique","mail_id": "unique"})
```

```python
unique_no_nan = lambda x: list(filter(None, pd.unique(x)))

df_train_pp.groupby(["InstanceNumber","BusinessActivity"]).agg({"applicant_id": unique_no_nan, "activity_id": unique_no_nan,"mail_id": unique_no_nan})
```

```python
unique_no_nan = lambda x: list(filter(None, pd.unique(x)))

df_train_pp.groupby(["InstanceNumber","BusinessActivity"]).agg(
    {"sale_order_id": unique_no_nan, "sale_order_line_id": unique_no_nan, "purchase_requisition_id": unique_no_nan,
     "purchase_requisition_line_id": unique_no_nan, "purchase_order_id": unique_no_nan,
     'purchase_order_line_id': unique_no_nan})
```
